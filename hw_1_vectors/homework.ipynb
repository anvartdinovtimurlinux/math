### Уровень 0:
Даны вектора x=(1,1), y=(2, 0) и z = (0,2)

Надо:
* Изобразить вектора на экране
* Изобразить точку x + y + z
* Найти угол между векторами x и y

```python
x = np.array([1, 1])
y = np.array([2, 0])
z = np.array([0, 2])

new_point = x + y + z

plt.figure()
plt.grid()
plt.plot(*x, 'o')
plt.plot(*y, 'o')
plt.plot(*z, 'o')
plt.plot(*new_point, 'x')
```

```python
def get_norm_vector(vector):
    return vector / np.linalg.norm(vector)

def get_angle(vector1, vector2): 
    vector1_norm = get_norm_vector(vector1) 
    vector2_norm = get_norm_vector(vector2) 
    return np.arccos(np.clip(np.dot(vector1_norm, vector2_norm), -1.0, 1.0)) 

alpha = get_angle(x, y)
print(f'Угол между векторами {x} и {y} равен {alpha} rad')
print(f'Или {alpha * 180 / np.pi}\u00B0')
```

### Уровень 1:
Даны дома из Бостона
* Выбрать любые n домов, образующих базис в пространстве признаков дома (все колонки, кроме medv)
* Написать код разложения дома по указанному базису

```python
boston_df = pd.read_csv('datasets_1379_2485_housing.csv')[['RM', 'LSTAT', 'PTRATIO']]
```

У нас три столбца в таблице, поэтому мы можем построить базис на значениях 3 домой.


```python
boston_ar = np.array(boston_df)

x = boston_ar[0]
y = boston_ar[1]
z = boston_ar[2]
basis = np.array([x, y, z]).transpose()
```

```python
houses_for_basis = []
for house in boston_ar:
    x = np.linalg.solve(basis, house)
    houses_for_basis.append({
        'a1': x[0],
        'a2': x[1],
        'a3': x[2],
    })

print(*houses_for_basis, sep="\n")
```
